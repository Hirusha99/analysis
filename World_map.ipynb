```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Initialize a map plot
fig, ax = plt.subplots(figsize=(12, 6))

# Set up the Basemap for a world map
m = Basemap(projection='cyl', llcrnrlat=-60, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution='c')
m.drawmapboundary(fill_color='white')  # Map boundary
m.fillcontinents(color='lightgray', lake_color='white')  # Land and lake colors
m.drawcoastlines()

# Define locations (latitude, longitude) and categories
locations = [
    (37.8, -122.4),   # USA (Explainable AI - Red)
    (52.5, 13.4),     # Germany (Nature Inspired - Orange)
    (28.6, 77.2),     # India (Multi-Agent - Blue)
    (-23.5, -46.6),   # Brazil (Machine Learning - Yellow)
    (-33.9, 151.2),   # Australia (Transfer Learning - Green)
    (55.8, 37.6)      # Russia (Deep Learning - Purple)
]

# Define the marker styles and colors
categories = [
    "Deep Learning", "Machine Learning", "Transfer Learning",
    "Multi-Agent", "Nature Inspired", "Explainable AI"
]
colors = ['purple', 'yellow', 'green', 'blue', 'orange', 'red']

# Plot each point
for (lat, lon), color in zip(locations, colors):
    m.scatter(lon, lat, color=color, s=100, label='_nolegend_', edgecolor='k', zorder=5)

# Add a legend
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label)
                   for color, label in zip(colors, categories)]
plt.legend(handles=legend_elements, loc='lower left', title="Categories")

# Add a title
plt.title("AI Topics Mapped Across the World")

# Show the plot
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the Natural Earth shapefile
world = gpd.read_file('/content/img/ne_110m_admin_0_countries.shp')  # Ensure path is correct!

# Step 2: Example data with updated species counts
data = {
    'BRK_A3': ['USA', 'IND', 'DEU', 'JPN', 'GBR', 'LKA', 'CAN', 'AUS', 'NLD', 'BGD',
               'CHN', 'ITA', 'VEN', 'OMN', 'MAR', 'TWN', 'MEX', 'NGA'],  # BRK_A3 country codes
    'species_count': [26, 21, 12, 8, 6, 3, 3, 3, 3, 2, 2, 2, 1, 1, 1, 1, 1, 1]  # Updated values
}
df = pd.DataFrame(data)

# Step 3: Merge with GeoDataFrame on 'BRK_A3'
world_data = world.merge(df, on='BRK_A3', how='left')

# Step 4: Plot the map
fig, ax = plt.subplots(figsize=(15, 10))

# Plot countries without data as light gray
world_data[world_data['species_count'].isna()].plot(
    ax=ax, color='lightgray', edgecolor='0.8', linewidth=0.5, label='No Data'
)

# Plot countries with data using species_count for coloring
world_data[~world_data['species_count'].isna()].plot(
    ax=ax, column='species_count', cmap='Reds', edgecolor='0.8', linewidth=0.5, legend=True
)

# Adjustments for overlapping country names and species counts
label_offsets = {
    'GBR': (0.5, 0),  # Adjust UK label position
    'DEU': (-0.5, 0),  # Adjust Germany label position
    'NLD': (0, 0.5)   # Adjust Netherlands label position
}

# Annotate countries with their names and species counts
for idx, row in world_data.dropna(subset=['species_count']).iterrows():
    country_name = row['NAME']
    species_count = row['species_count']
    x, y = row.geometry.centroid.x, row.geometry.centroid.y

    # Apply offsets for specific countries
    if country_name in label_offsets:
        offset_x, offset_y = label_offsets[country_name]
        x += offset_x
        y += offset_y

    # Annotate the map with country name and species count
    ax.text(x, y, f'{country_name}: {species_count}', fontsize=8, ha='center', color='black')

# Add arrows for UK, Germany, and Netherlands
arrow_params = {
    'GBR': {'x': 0.5, 'y': 0, 'text': 'UK: 6', 'text_offset': (0.7, 0)},
    'DEU': {'x': -0.5, 'y': 0, 'text': 'Germany: 12', 'text_offset': (-0.7, 0)},
    'NLD': {'x': 0, 'y': 0.5, 'text': 'Netherlands: 3', 'text_offset': (0, 0.7)}
}

for country, params in arrow_params.items():
    # Find the centroid of the country
    country_row = world_data[world_data['NAME'] == country]
    x, y = country_row.geometry.centroid.x.values[0], country_row.geometry.centroid.y.values[0]

    # Adjust x and y with offsets
    x += params['x']
    y += params['y']

    # Add arrow
    ax.annotate('', xy=(x + params['x'], y + params['y']), xytext=(x, y),
                arrowprops=dict(facecolor='black', edgecolor='black', arrowstyle='->', lw=1))

    # Add text label
    ax.text(x + params['text_offset'][0], y + params['text_offset'][1], params['text'],
            fontsize=8, ha='center', color='black')

# Add title and legend
plt.title("Number of Species Occurrences by Country", fontsize=15)
plt.axis('off')  # Remove axes
plt.legend(loc='lower left', title="Legend", frameon=False)

# Show the map
plt.show()
```
